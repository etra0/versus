```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('./src/data.json')
```

```python
df.views = df.views.str.replace('M views', '*1e6')
df.views = df.views.str.replace('B views', '*1e9')
df.views = df.views.str.replace('K views', '1e3')
```

```python
df_parsed = (df
 .assign(viewsNumber = df.views.apply(eval))
)
```

```python
g = sns.FacetGrid(df_parsed, col='band', col_wrap=3, sharex=False)
g.map(plt.hist, 'viewsNumber')
```

```python
plt.hist(df_parsed.viewsNumber, bins=20)
```
